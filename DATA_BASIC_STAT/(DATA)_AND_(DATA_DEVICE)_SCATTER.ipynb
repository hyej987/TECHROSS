# 산점도를 통해 변수들 간의 상관성 파악

- 산점도로 표현하기에는 열의 수가 굉장히 많으므로, HEATMAP으로 상관성을 파악하였다.

- 숫자를 모두 반영하는 것은 비효율적이기 때문에, 상관계수가 0.5가 넘어가면 1이라고 판정하였고, 0.5 이하이면 0이라고 판정하였다.

# 패키지 로딩

```python
from google.colab import drive
drive.mount('/content/gdrive')
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

# DATA_DEVICE_DATA_OPTIME의 산점도

- DATA를 중심으로 DATA_DEVICE 테이블을 LEFT JOIN하고, 이를 DATA_DEVICE_DATA라 하면, DATA_DEVICE_DATA에 또다시 OPTIME 데이터 셋을 LEFT JOIN한 테이블인 DATA_DEVICE_DATA_OPTIME 테이블이다.

- 이 때, DATA_DEVICE_DATA_OPTIME 테이블의 산점도를 그려보았다.

- 데이터가 크므로 참고용으로 파악한다.

```python
DATA_DEVICE_DATA_OPTIME = pd.read_csv('/content/gdrive/MyDrive/DB/DATA_DEVICE_DATA_OPTIME.csv')
```

```python
corr=DATA_DEVICE_DATA_OPTIME.corr()

data = []
for i in range(len(corr)) : 
  data.append(pd.Series(corr.iloc[i]>0.5))

data = pd.DataFrame(data)

plt.figure(figsize = (40,40))
sns.heatmap(data, annot=True)
```
