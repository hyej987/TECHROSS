- CITY 테이블에는 총 4개의 범주형 변수가 존재
- CITY, COUNTRY, ISO2, ISO3

# 패키지로딩

# 데이터 로딩

```python
EVENT = pd.read_csv('/content/gdrive/MyDrive/DB/CITY.csv')
list_event = ['CITY', 'COUNTRY', 'ISO2', 'ISO3']
for i in range(len(list_event)) :
 EVENT[f'{list_event[i]}']=EVENT[f'{list_event[i]}'].astype(str)
```

# 100% 플랏

```python
list_event_data = []

for i in range(len(list_event)) :
  list_event_data.append(pd.DataFrame(EVENT[f"{list_event[i]}"].dropna().value_counts()))

for i in range(len(list_event)) :
  list_event_data[i] = list_event_data[i][list_event_data[i].index != 'nan']

for i in range(len(list_event)) :
  plt.figure(figsize = (30,10))
  plt.bar(list_event_data[i].index, list_event_data[i][f'{list_event[i]}'])
  plt.xticks(rotation = 90)
  plt.title('CITY TABLE' +"     " +f'{list_event[i]}',fontsize = 20, pad = 20)
  plt.savefig('/content/gdrive/MyDrive/TECHCROSS 자료파일/범주형 자료 분석(빈도분석 바차트)/CITY/' + f'{list_event[i]}' + ".png")
```

# 90% 플랏

```python

list_event_data_90 = []

for i in range(len(list_event)) :
  list_event_data_90.append(list_event_data[i].reset_index().stb.freq(['index'], value = f'{list_event[i]}', thresh = 90)[["index","percent"]])

for i in range(len(list_event)) :
  plt.figure(figsize = (30,10))
  plt.bar(list_event_data_90[i].iloc[:,0], list_event_data_90[i].iloc[:,1])
  plt.xticks(rotation = 90)
  plt.title('CITY TABLE' +"     " +f'{list_event[i]}'+ "      " + "90%",fontsize = 20, pad = 20)
  plt.savefig('/content/gdrive/MyDrive/TECHCROSS 자료파일/범주형 자료 분석(빈도분석 바차트)/CITY/' + f'{list_event[i]}' +"   90%" +".png")
```
