# data , drop duplicate, drop unknowns

```python
data = pd.read_csv('../csv/DATA.csv')
```

```python
data_index = data['DATA_INDEX']
```

```python
data_copy = data.copy()
data.drop('DATA_INDEX', axis=1, inplace=True)
```

```python
data = data.drop_duplicates()
```

```python
drop_index = list(data[(data['SHIP_ID'] == 'Unknown') | (data['SHIP_ID'] == 'UnkownShip')].index)
```

```python
data = data.drop(drop_index)
```

```python
onlist = list(data.columns)
```

```python
data.shape
```

```python
data['index']= list(data.index)
data_copy['index']= list(data_copy.index)
```

```python
data = pd.merge(data, data_copy, how='inner', on=onlist)
```

```python
data.to_csv('data_trim.csv', mode='w')
```

# Mcode data 

```python
# data minute 

data['DATA_TIME'] = data['DATA_TIME'].astype('datetime64')

year = []
month = []
day = []
hour = []
minute = []

make =[year, month, day, hour, minute]

for i in range(len(data['DATA_TIME'])):
    year.append(data.loc[i, 'DATA_TIME'].year)
    month.append(data.loc[i, 'DATA_TIME'].month)
    day.append(data.loc[i, 'DATA_TIME'].day)
    hour.append(data.loc[i, 'DATA_TIME'].hour)
    minute.append(data.loc[i, 'DATA_TIME'].minute)
    
data['year'] = year
data['month'] = month
data['day'] = day
data['hour'] = hour
data['minute'] = minute
```

# Fault code

```python
fault_code = pd.get_dummies(nevt['FAULT_CODE'])
fault_codelist = list(fault_code.columns)

for i in range(len(fault_codelist)):
    fault_codelist[i] = ('faultcode_' + f'{fault_codelist[i]}')
```

```python
fault_code.columns = fault_codelist
```

```python
nevt = pd.concat([nevt, fault_code], axis=1)
nevt.head()
```

```python
origin = nevt.groupby(['SHIP_ID', 'SECTION', 'OP_INDEX', 'year', 'month', 'day', 'hour', 'minute'])['0.0'].sum().reset_index().iloc[:, :8]

for i in fault_codelist:
    temp = nevt.groupby(['SHIP_ID', 'SECTION', 'OP_INDEX', 'year', 'month', 'day', 'hour', 'minute'])[i].sum().reset_index().iloc[:, -1]
    origin[f'{i}'] = temp
```

```python
origin.head()
```

# fault code data 

```python
data_evt_code = pd.merge(data_mcode, origin, how='left', on=['SHIP_ID', 'SECTION', 'OP_INDEX', 'year', 'month', 'day', 'hour', 'minute'])
```

```python
data_evt_code.drop('index', axis=1, inplace=True)
data_evt_code.to_csv('data_evt_code.csv', mode='w')
```
