# ECU 유닛 1:8 모두 0인 것을 제외하면 CURRENT는 유닛 1~8 합, VOLTAGE는 평균 여부 확인

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
ecu = pd.read_csv(('../../../Work/techcross/csv/ECU.csv'))
```

```python
ecu.shape
```

## VOLTAGE CHECK

```python
#V1:8 전부 0인것은 제외
vcon = (ecu['V1'] != 0) | (ecu['V2'] != 0) | (ecu['V3'] != 0) | (ecu['V4'] != 0) | (ecu['V5'] != 0) | (ecu['V6'] != 0) | (ecu['V7'] != 0) | (ecu['V8'] != 0)
```

```python
ecuvcon = ecu[vcon]
```

```python
#0을 널값으로 변환, 평균내어 첫째자리 까지 반올림
ecuvcon = ecuvcon.replace(0, np.NaN)
ecuvcon['Vavg'] = ecuvcon.iloc[:, 4:11].mean(axis=1).round(1)
ecuvcon['일치여부1'] = ecuvcon['VOLTAGE'] == ecuvcon['Vavg']
ecuvcon
```

```python
#첫째자리까지 반올림 했을 때 불일치 하는 것들
ecuvcon[ecuvcon['일치여부1'] == False]
```

```python
#둘째자리 까지 반올림한 후 오차범위 0.1로, 첫째자리까지 반올림하여 판단
ecuvcon['Vavg'] = ecuvcon.iloc[:, 4:11].mean(axis=1).round(2)
ecuvcon['일치여부2'] = (ecuvcon['VOLTAGE'] <= (ecuvcon['Vavg'] + 0.1).round(1)) & (ecuvcon['VOLTAGE'] >= (ecuvcon['Vavg'] - 0.1).round(1))
ecuvcon[ecuvcon['일치여부2'] == False]
```

```python
ecuvcon[ecuvcon['일치여부2'] == False]['SHIP_ID'].value_counts()
```

```python
#유닛1:8이 모두 0이고 VOLTAGE도 0이어서 NULL인지, 정상치인지 판단 애매
vcon_nnull = (ecu['V1'] == 0) & (ecu['V2'] == 0) & (ecu['V3'] == 0) & (ecu['V4'] == 0) & (ecu['V5'] == 0) & (ecu['V6'] == 0) & (ecu['V7'] == 0) & (ecu['V8'] == 0) & (ecu['VOLTAGE'] == 0)
ecu[vcon_nnull]
```

## CURRENT CEHCK

```python
#C1:8 전부 0인것은 제외
ccon = (ecu['C1'] != 0) | (ecu['C2'] != 0) | (ecu['C3'] != 0) | (ecu['C4'] != 0) | (ecu['C5'] != 0) | (ecu['C6'] != 0) | (ecu['C7'] != 0) | (ecu['C8'] != 0)
ecuccon = ecu[ccon]
ecuccon['Csum'] = ecuccon.iloc[:, 12:20].sum(axis=1)
ecuccon['일치여부1'] = ecuccon['CURRENT'] == ecuccon['Csum']
ecuccon
```

```python
ecuccon[ecuccon['일치여부1'] == False]
```

```python
ecuccon['일치여부2'] = (ecuccon['CURRENT'] <= (ecuccon['Csum'] + 100)) & (ecuccon['CURRENT'] >= (ecuccon['Csum'] - 100))
ecuccon[ecuccon['일치여부2'] == False]
```
