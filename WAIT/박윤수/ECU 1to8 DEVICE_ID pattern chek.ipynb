```python
#호선, 구역, 장치. 데이터번호로 재배열
ecu = ecu.sort_values(by=['SHIP_ID', 'SECTION', 'DEVICE_ID', 'D_INDEX'])
ecu = ecu[['SHIP_ID', 'SECTION', 'DEVICE_ID', 'D_INDEX', 'CONTROL', 'V1', 'V2', 'V3', 'V4', 'V5', 'V6', 'V7', 'V8', 'C1', 'C2', 'C3', 'C4', 'C5', 'C6', 'C7', 'C8', 'TEMP', 'CONTROL_CURRENT', 'CURRENT', 'VOLTAGE']]
```

```python
SHIP_ID=ecu['SHIP_ID'].unique()
SHIP_ID
```

```python
#호선별, , SECTION별 DEVICE_ID별 C1:C8 on off 여부 확인해보기
SHIP_ID=ecu['SHIP_ID'].unique()
SHIP_ID

for i in range(35):
    SHIP_ID[i]=ecu.loc[ecu['SHIP_ID'] == SHIP_ID[i],:]

SHIP_ID_BIT = []
SHIP_DEVICE_ID = []
C = []
BIT_CONTROL = []
BIT = []
for i in range(35) :
    
    SHIP_DEVICE_ID.append(pd.concat([SHIP_ID[i][['SECTION', 'DEVICE_ID']], SHIP_ID[i].iloc[:,13:21]],axis = 1))
    C.append(SHIP_DEVICE_ID[i].iloc[:,1:])
    C[i] = C[i] != 0 
    C[i]=C[i].astype(int)
    C[i]=C[i]['C1'].astype(str) + C[i]['C2'].astype(str) + C[i]['C3'].astype(str) + C[i]['C4'].astype(str) + C[i]['C5'].astype(str) + C[i]['C6'].astype(str) + C[i]['C7'].astype(str) + C[i]['C8'].astype(str)
    BIT_CONTROL.append(pd.concat([SHIP_ID[i][['SECTION', 'DEVICE_ID']],C[i]], axis = 1))
    BIT.append(BIT_CONTROL[i].drop_duplicates())
    SHIP_ID_BIT.append(pd.DataFrame(BIT[i].sort_values(by = ['SECTION', 'DEVICE_ID'])))
```
