# ECU SHIP_ID, SECTION, LINE 별로 데이터 분리

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#파일 불러오기
ecu = pd.read_csv('../../../Work/techcross/csv/ECU.csv')
opt_d_dd = pd.read_csv('../../../Work/techcross/0318/DATA_DEVICE_DATA_OPTIME.csv')
```

```python
ecu = opt_d_dd[opt_d_dd['DEVICETYPE'] == 1]
```

```python
ecu.columns
```

```python
#ecu 인덱싱 및 정리
ecu = ecu[['SHIP_ID', 'SECTION', 'LINE', 'DEVICE_ID', 'OP_INDEX', 'DATA_INDEX', 'D_INDEX', 'ECU_V1',
       'ECU_V2', 'ECU_V3', 'ECU_V4', 'ECU_V5', 'ECU_V6', 'ECU_V7', 'ECU_V8',
       'ECU_C1', 'ECU_C2', 'ECU_C3', 'ECU_C4', 'ECU_C5', 'ECU_C6', 'ECU_C7',
       'ECU_C8', 'ECU_TEMP', 'ECU_CURRENT', 'ECU_VOLTAGE',
       'ECU_CONTROL_CURRENT']]
ecu.sort_values(by=['SHIP_ID', 'SECTION', 'LINE', 'DEVICE_ID', 'OP_INDEX', 'DATA_INDEX', 'D_INDEX'], inplace=True)
ecu
```

```python
#배 id 리스트
temp = ecu['SHIP_ID'].unique()
temp.tolist()
```

```python
ecu['key'] = ecu['SHIP_ID'] + '_' + ecu['SECTION'].astype(str) + '_' + ecu['LINE'].astype(str)
```

```python
ecukeylist = ecu['key'].unique().tolist()
```

```python
temlist = []
```

```python
for _ in range(0, len(ecukeylist)):
    temlist.append(f'ecutemp{_}')
```

```python
temlist
```

```python
for _ in range(0, len(ecukeylist)):
    temlist[_] = ecu[ecu['key'] == ecukeylist[_]]
    temlist[_].to_csv(f'ecutemp{_}.csv', mode='w')
```

```python
temlist[_]['C_SUM'] = temlist[_]['ECU_C1'] + temlist[_]['ECU_C2'] + temlist[_]['ECU_C3'] + temlist[_]['ECU_C4'] + temlist[_]['ECU_C5'] + temlist[_]['ECU_C6'] + temlist[_]['ECU_C7'] + temlist[_]['ECU_C8']
```
