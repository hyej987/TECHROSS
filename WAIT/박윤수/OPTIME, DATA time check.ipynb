```python
data = pd.read_csv('../../../Work/techcross/csv/DATA.csv')
opt = pd.read_csv('../../../Work/techcross/csv/OPTIME.csv')
```

```python
opt['key'] = opt['SHIP_ID'] + '_' + opt['SECTION'].astype(str) + '_' + opt['OP_INDEX'].astype(str)
opt['key'].value_counts()
```
