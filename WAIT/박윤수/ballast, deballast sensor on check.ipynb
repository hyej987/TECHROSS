```python
#Ballast중인 Data
ballast = opt_d_dd[opt_d_dd['OP_TYPE'] == 1]
ballast.drop('Unnamed: 0', axis=1, inplace=True)
ballast
```

# ballast 중 장치 동작 확인

```python
#ballast 중 ANU 동작 확인(동작하지 않아야 함)
temp = ballast[['ANU_ALARM', 'ANU_STATE', 'ANU_CONTROL', 'ANU_AMOUNT1', 'ANU_AMOUNT2', 'ANU', 'ANU_AMOUNT']]
temp = temp.isnull()
print(temp[temp['ANU_ALARM'] == False])
print(temp[temp['ANU_STATE'] == False])
print(temp[temp['ANU_CONTROL'] == False])
print(temp[temp['ANU_AMOUNT1'] == False])
print(temp[temp['ANU_AMOUNT2'] == False])
print(temp[temp['ANU'] != 0])
print(temp[temp['ANU_AMOUNT'] != 0])
```

```python
#ballast 중 TSU 동작 확인(회의록대로는 동작하지 않아야 함)
temp2 = ballast[['SHIP_ID', 'SECTION', 'LINE', 'DEVICE_ID', 'TSU_STATE', 'TSU_CONTROL', 'TSU_BP1_CLX1', 'TSU_BP1_CLX2', 'TSU_BP2_CLX1', 'TSU_BP2_CLX2', 'TSU_BP3_CLX1', 'TSU_BP3_CLX2', 'TRO', 'TRO_COUNT', 'TRO_AVG']]
```

```python
#TSU_BP1_CLX1:TSU_BP3_CLX2중 하나가 null이 아니면서 0도 아닌 값
condi1 = ((temp2['TSU_BP1_CLX1'].notnull()) & (temp2['TSU_BP1_CLX1'] != 0))
condi2 = ((temp2['TSU_BP1_CLX2'].notnull()) & (temp2['TSU_BP1_CLX2'] != 0))
condi3 = ((temp2['TSU_BP2_CLX1'].notnull()) & (temp2['TSU_BP2_CLX1'] != 0))
condi4 = ((temp2['TSU_BP2_CLX2'].notnull()) & (temp2['TSU_BP2_CLX2'] != 0))
condi5 = ((temp2['TSU_BP3_CLX1'].notnull()) & (temp2['TSU_BP3_CLX1'] != 0))
condi6 = ((temp2['TSU_BP3_CLX1'].notnull()) & (temp2['TSU_BP3_CLX2'] != 0))

temp2[condi1 | condi2 | condi3 | condi4 | condi5 | condi6].to_csv('tsu ballast 동작확인.csv', mode='w')
```
