```python
#통합 칼럼 만든 후, 필터링

msg_nerr['all'] = msg_nerr['MESSAGE'].astype(str) + '_' + msg_nerr['DEVICE_NAME'].astype(str) + '_' + msg_nerr['TS_URL'].astype(str)

filt_list = ['ECU', 'TSU', 'PMU', 'ANU', 'TRO', 'CSU', 'FMU', 'STS', 'FTS', 'CLX', 'DTU', 'Drain Tank',]

for i in range(0, len(filt_list)):
    globals()[filt_list[i]] = msg_nerr[msg_nerr['all'].str.contains(filt_list[i])]
    globals()[filt_list[i] + 'errcode'] = globals()[filt_list[i]]['M_CODE'].unique()
```

```python
checknlist = []
checknlistsum = 0

for i in range(0, len(filt_list)):
    print(f'{filt_list[i]} ERR_CODE ' + str(len(globals()[filt_list[i]])) + '개')
    checknlist.append(len(globals()[filt_list[i]]))
    checknlistsum += len(globals()[filt_list[i]])
    
print(f'\n검출된 ERR_CODE의 총 합 개수 = {checknlistsum}')
```
