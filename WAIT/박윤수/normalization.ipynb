```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from sklearn.datasets import load_iris
from sklearn.preprocessing import MinMaxScaler
warnings.filterwarnings('ignore')
```

```python
opt_d_dd = pd.read_csv('../../../Work/techcross/0318/DATA_DEVICE_DATA_OPTIME.csv')
```

# 정규화

```python
opt_d_dd.columns
```

```python
#정규화 할 리스트 인덱싱
list = ['ECU_V1', 'ECU_V2', 'ECU_V3', 'ECU_V4', 'ECU_V5', 'ECU_V6', 'ECU_V7', 'ECU_V8', 'ECU_C1', 'ECU_C2', 'ECU_C3', 'ECU_C4', 'ECU_C5', 'ECU_C6', 'ECU_C7', 'ECU_C8', 'ECU_TEMP', 'ECU_CURRENT', 'ECU_VOLTAGE', 'ECU_CONTROL_CURRENT', 'ANU_AMOUNT1', 'ANU_AMOUNT2', 'TSU_BP1_CLX1', 'TSU_BP1_CLX2', 'TSU_BP2_CLX1', 'TSU_BP2_CLX2', 'TSU_BP3_CLX1', 'TSU_BP3_CLX2', 'PMU_V1', 'PMU_V2', 'PMU_V3', 'PMU_C1', 'PMU_C2', 'PMU_C3', 'PMU_PSUM', 'CSU', 'STS', 'FTS', 'FMU', 'TRO', 'ANU', 'RATE', 'CURRENT', 'VOLTAGE', 'VOLUME', 'TRO_AVG', 'ANU_AMOUNT', 'TOTAL_POWER']
optddd_normal = opt_d_dd[list]

# MinMaxcaler객체 생성
scaler = MinMaxScaler()

# MinMaxScaler로 데이터 셋 변환 .fit() 과 .transform() 호출.  
scaler.fit(optddd_normal)
optddd_normal_scaled = scaler.transform(optddd_normal)

# transform()시 scale 변환된 데이터 셋이 numpy ndarry로 반환되어 이를 DataFrame으로 변환
optddd_normal_df_scaled = pd.DataFrame(data=optddd_normal_scaled, columns=list)

print('정규화 최소 값')
print(optddd_normal_df_scaled.min())
print('\n정규화 최대 값')
print(optddd_normal_df_scaled.max())
```
