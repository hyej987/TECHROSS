# 오피타임 + 이벤트 로그레벨 별로 통계 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
import sidetable
warnings.filterwarnings('ignore')
```

```python
#필요한 파일 불러오기

evt = pd.read_csv('../../../Work/techcross/csv/EVENT.csv')
opt = pd.read_csv('../../../Work/techcross/csv/OPTIME.csv')
opt_e = pd.read_csv('../../../Work/techcross/csv/OPTIME_EVENT.csv')
```

```python
#병합

print(evt.shape)
print(opt.shape)
print(opt_e.shape)

opt_evt = pd.merge(opt_e, evt, how='left', on=['SHIP_ID', 'SECTION', 'EVENT_INDEX'])
print(opt_evt.shape)

evt_opt = pd.merge(opt_evt, opt, how='left', on=['SHIP_ID', 'SECTION', 'OP_INDEX'])
print(evt_opt.shape)
```

## 기본정보

```python
#정보

evt_opt.info()
```

```python
print('결측값 개수와 비율\n')
evt_opt.stb.missing(style=True)
```

## 수치형 자료들 통계

```python
#describe

pd.options.display.float_format = '{:.2f}'.format
describe_list = ['RUNNING_TIME', 'VOLUME', 'TRO_AVG', 'ANU_AMOUNT', 'ALARM_COUNT', 'FAULT_COUNT', 'TOTAL_POWER']

log_lv0 = evt_opt[evt_opt['LOG_LEVEL'] == 0]
log_lv1 = evt_opt[evt_opt['LOG_LEVEL'] == 1]
log_lv2 = evt_opt[evt_opt['LOG_LEVEL'] == 2]
```

```python
print("LOG_LEVEL이 0일 때, 수치형 자료들 통계\n")
pd.DataFrame(log_lv0[describe_list].describe())
```

```python
print("LOG_LEVEL이 1일 때, 수치형 자료들 통계\n")
pd.DataFrame(log_lv1[describe_list].describe())
```

```python
print("LOG_LEVEL이 2일 때, 수치형 자료들 통계\n")
pd.DataFrame(log_lv2[describe_list].describe())
```
