```python
from google.colab import drive
drive.mount('/content/gdrive')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
EVENT = pd.read_csv('/content/gdrive/MyDrive/DB/EVENT.csv')
```

```python
DEVICE_NAME=pd.DataFrame(EVENT['DEVICE_NAME'].value_counts())
plt.figure(figsize = (30,10))
plt.bar(DEVICE_NAME.index, DEVICE_NAME['DEVICE_NAME'])
plt.xticks(rotation = 90)
plt.title('EVENT TABLE DEVICE_NAME', fontsize = 20, pad = 20)
```
